# Catastrophe Model

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate 1000 samples of a negative binomial distribution
r = 5
p = 0.2
n_samples = 1_000_000
samples = np.random.negative_binomial(r, p, n_samples)
```

```python
# Simulate zero-inflated negative binomial distribution
zero_inflation_prob = 0.3  # Probability of zero inflation
zero_inflated_samples = np.where(np.random.rand(n_samples) < zero_inflation_prob, 0, samples)

fig, ax = plt.subplots(1, 2, figsize=(12, 5), dpi=300)

# Plot the histogram of the samples
ax[0].hist(samples, bins=30, edgecolor='k', alpha=0.7)
ax[0].set_title('Negative Binomial Distribution (r=5, p=0.2)')
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Frequency')

# Plot the histogram of the zero-inflated samples
ax[1].hist(zero_inflated_samples, bins=30, edgecolor='k', alpha=0.7)
ax[1].set_title('Zero-Inflated Negative Binomial (zero-inflation=0.3)')
ax[1].set_xlabel('Value')
ax[1].set_ylabel('Frequency')

plt.suptitle('Negative Binomial vs Zero-Inflated Negative Binomial', fontsize=16, fontweight='bold')
plt.tight_layout()

plt.show()
```
